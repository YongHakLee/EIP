튜플(tuple)은 몇 가지 점을 제외하면 리스트와 거의 비슷하다. 리스트와 다른 점은 다음과 같다.

- 리스트는 [], 튜플은 ()으로 둘러싼다.
- 리스트는 요솟값의 생성, 삭제, 수정이 가능하지만, 튜플은 요솟값을 바꿀 수 없다.


### 튜플은 어떻게 만들까?


```python
t1 = ()
t2 = (1,)
t3 = (1, 2, 3)
t4 = 1, 2, 3
t5 = ("a", "b", ("ab", "cd"))
```

모습은 리스트와 거의 비슷하지만, 튜플에서는 리스트와 다른 2가지 차이점을 찾아볼 수 있다. t2 = (1,)처럼 단지 1개의 요소만을 가질 때는 요소 뒤에 쉼표(,)를 반드시 붙여야 한다는 것과 t4 = 1, 2, 3처럼 소괄호(())를 생략해도 된다는 점이다.

튜플과 리스트는 비슷해 보이지만, 구별해서 사용하는 것이 좋다. 가장 큰 차이는 요솟값을 변경할 수 있는지 여부이다. 리스트의 요솟값은 변경 가능하고, 튜플의 요솟값은 변경 불가능하다.

프로그램 실행 중 요솟값이 변하지 않기를 바란다면 튜플을 사용하자. 반대로 값을 수시로 변경해야 한다면 리스트를 사용하자. 실제 프로그램에서는 값이 자주 변경되는 경우가 많아 튜플보다 리스트를 더 자주 사용한다.


### 튜플의 요솟값을 지우거나 변경하려고 할 때


#### 튜플의 요솟값을 삭제하려 할 때


```python
t1 = (1, 2, "a", "b")
del t1[0]
```

#### 2. 튜플 요솟값을 변경하려 할 때


```python
t1 = (1, 2, "a", "b")
t1[0] = "c"
```

### 튜플 다루기


#### 인덱싱하기


```python
t1 = (1, 2, "a", "b")
t1[0]
```

#### 슬라이싱하기


```python
t1 = (1, 2, "a", "b")
t1[1:]
```

#### 튜플 더하기 (연결)


```python
t1 = (1, 2, "a", "b")
t2 = (3, 4)
t3 = t1 + t2
t3
```

t1, t2 튜플의 요솟값이 바뀌는 것이 아니라, 더해서 새로운 튜플 t3을 생성한 것이다.


#### 튜플 곱하기 (반복)


```python
t2 = (3, 4)
t3 = t2 * 3
t3
```

### 튜플 길이 구하기


```python
t1 = (1, 2, "a", "b")
len(t1)
```
