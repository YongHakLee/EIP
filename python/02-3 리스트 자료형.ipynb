# 02. 자료형


## 02-3 리스트 자료형


### 리스트는 어떻게 만들고 사용할까?


```python
a = []
b = [1, 2, 3]
c = ["Life", "is", "too", "short"]
d = [1, 2, "Life", "is"]
e = [1, 2, ["Life", "is"]]
f = list()
```

리스트는 a처럼 아무것도 포함하지 않아 비어 있는 리스트([])일 수도 있고, b처럼 숫자를 요솟값으로 가질 수도 있으며, c처럼 문자열을 요솟값으로 가질 수도 있다. 또한 d처럼 숫자와 문자열을 함께 요솟값으로 가질 수도 있고, e처럼 리스트 자체를 요솟값으로 가질 수도 있다. 즉, 리스트 안에는 어떠한 자료형도 포함할 수 있다.


### 리스트의 인덱싱과 슬라이싱


#### 리스트의 인덱싱


```python
a = [1, 2, 3]
a[0]
```

```python
a[0] + a[2]
```

```python
a[-1]
```

```python
a = [1, 2, 3, ["a", "b", "c"]]
```

```python
print(a[0])
print(a[-1])
print(a[-1][0])
```

#### 리스트의 슬라이싱


```python
a = [1, 2, 3, 4, 5]
a[0:2]
```

```python
a = "12345"
a[0:2]
```

```python
a = [1, 2, 3, 4, 5]
b = a[:2]
c = a[2:]
print(a)
print(b)
print(c)
```

### 리스트 연산하기


#### 리스트 더하기(+)


```python
a = [1, 2, 3]
b = [4, 5, 6]
a + b
```

#### 리스트 반복하기(\*)


```python
a = [1, 2, 3]
a * 3
```

#### 리스트 길이 구하기


```python
a = [1, 2, 3]
len(a)
```

### 리스트의 수정과 삭제


#### 리스트의 값 수정하기


```python
a = [1, 2, 3]
a[2] = 4
a
```

#### del로 리스트 요소 삭제하기


```python
a = [1, 2, 3]
del a[1]
a
```

```python
a = [1, 2, 3, 4, 5]
del a[2:]
a
```

### 리스트 관련 함수


#### 리스트에 요소 추가하기 - append


```python
a = [1, 2, 3]
a.append(4)
a
```

```python
a.append([5, 6])
a
```

#### 리스트 정렬 - sort


```python
a = [1, 4, 3, 2]
a.sort()
a
```

#### 리스트 뒤집기 - reverse


현재의 리스트를 그대로 거꾸로 뒤집는다.


```python
a = ["a", "c", "b"]
a.reverse()
a
```

#### 인덱스 반환 - index


리스트에 x 값이 있으면 x의 인덱스 값(위칫값)을 반환한다.


```python
a = [1, 2, 3]
print(a.index(3))
print(a.index(1))
```

```python
a.index(0)
```

#### 리스트에 요소 삽입 - insert


리스트의 a번째 위치에 b를 삽입하는 함수이다.


```python
a = [1, 2, 3]
a.insert(0, 4)
a
```

```python
a.insert(3, 5)
a
```

#### 리스트 요소 제거 - remove


리스트에서 첫 번째로 나오는 x를 삭제하는 함수이다.


```python
a = [1, 2, 3, 1, 2, 3]
a.remove(3)
a
```

```python
a.remove(3)
a
```

#### 리스트 요소 꺼내기 - pop


리스트의 맨 마지막 요소를 반환하고 그 요소는 삭제한다.


```python
a = [1, 2, 3]
a.pop()
```

#### 리스트에 포함된 요소 x의 개수 세기 - count


리스트 안에 x가 몇 개 있는지 조사하여 그 개수를 반환하는 함수이다.


```python
a = [1, 2, 3, 1]
a.count(1)
```

#### 리스트 확장 - extend


extend(x)에서 x에는 리스트만 올 수 있으며 원래의 a 리스트에 x 리스트를 더하게 된다.


```python
a = [1, 2, 3]
a.extend([4, 5])
a
```

```python
b = [6, 7]
a.extend(b)
a
```
